# KMeans
1. Select the Number of Clusters k
2. Select k Points at Random or from data points as centroids 
3. Make k Clusters based on min distance of each point with centroids. 
4. Compute New Centroid of Each Cluster(mean of each cluster)
5. Assess the Quality of Each Cluster
6. Repeat 3-5

```python
import random 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
df = pd.read_csv("Cust_Segmentation.csv")
df.head(10)
```

```python
#Address is not useful, we can put it a side. 
df = df.drop('Address', axis=1)
df.head(10)
```

```python
#Data_preprocessing
from sklearn.preprocessing import StandardScaler
x = df.values[:,1:] #exclude customer Id
x = np.nan_to_num(x) 
x = StandardScaler().fit_transform(x)
x[0:10]
```

```python
#Train the model
from sklearn.cluster import KMeans
cluster_Number = 4
k_means = KMeans(init = "k-means++", n_clusters = cluster_Number, n_init = 10)
k_means.fit(x)
labels = k_means.labels_
print(labels)
```

```python
#add a column to df
df["Cluster"] = labels
df.head(10)
```

```python
cdf = df.groupby('Cluster').mean()
cdf.drop('Customer Id', axis=1)
```
